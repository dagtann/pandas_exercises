### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
URL = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/"
URL_DATA = URL + "chipotle.tsv"
chipo = pd.read_csv(URL_DATA, sep="\t")
```

### Step 4. How many products cost more than $10.00?

```python
# Check item_price data quality
chipo.head()
chipo.info()
chipo["item_price_num"] = pd.Series([float(p[1:-1]) for p in chipo["item_price"]]) # dtype
any(chipo["item_price_num"].isnull()) # Missing data?
chipo["item_price_num"].describe()
any(chipo["item_name"].duplicated()) # Duplicate entries?
unique_chipo = chipo.drop_duplicates(["item_name"])
any(unique_chipo["quantity"] > 1)  # any multi-item orders? drop b/c can't tell $10 item
unique_chipo = unique_chipo[unique_chipo["quantity"] == 1]

# Count items more expensive than $10.0
#sum([p > 10.0 for p in chipo["item_price_num"]])
unique_chipo["item_price_num"][unique_chipo["item_price_num"] > 10].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[["item_name", "item_price"]]
```

### Step 6. Sort by the name of the item

```python
chipo[["item_name", "item_price"]].sort_values(by = "item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo["quantity"][chipo["item_price_num"] == chipo["item_price_num"].max()].sum()
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
chipo["quantity"][chipo["item_name"] == "Veggie Salad Bowl"].sum()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
chipo["order_id"][(chipo["item_name"] == "Canned Soda") & (chipo["quantity"] > 1)].count()
```
